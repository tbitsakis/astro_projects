# Distributions
In this notebook we are going to examine various statistical distributions. We are going to use scipy.stats

```python
# Importing Libraries
import numpy as np
import scipy
from scipy import stats
import matplotlib.pyplot as plt
```

- Poisson Distribution
The probability of $\nu$ counts in any definite interval:

$$
P(\nu|\mu) = \frac{\mu^\nu e^{-\mu}}{\nu!}
$$

where $\mu$ is the mean of the distribution.

Now we are going to plot some poissonian distributions.

```python
poisson = stats.poisson

mu_values = [5,10,15]
linestyles = ['-','--',':']

for mu,ls in zip(mu_values,linestyles):
    p = poisson(mu)
    x = np.arange(p.ppf(0.00001),p.ppf(0.99999)) # this is to create a range of x's that are found in the poisson range
    
    plt.plot(x,p.pmf(x),ls=ls,color='black',drawstyle='steps-mid',label=r'$\mu=$%i'%mu)
    
plt.xlabel(r'$\nu$')
plt.ylabel(r'$P(\nu|\mu)$')
plt.title('Poisson Distribution')
plt.legend()
plt.show()
```

The Poisson doesn't look like a Gaussian (normal) distribution. In astronomy when we have a large number of counts, the photon-distr. approximates a Gaussian, however the $\sigma=\sqrt{N}$.

```python
mu = 5
p_p = stats.poisson(mu)
p_g = stats.norm(mu,mu**0.5)
x = np.arange(p_p.ppf(0.0001),p_p.ppf(0.9999))

plt.plot(x,p_p.pmf(x),ls='-',color='black',drawstyle='steps-mid',label='Poisson')
plt.plot(x,p_g.pdf(x),ls=':',color='black',label='Gaussian')
plt.xlabel(r'$\nu$')
plt.ylabel(r'$P(\nu|\mu)$')
plt.legend()
plt.show()
```

We see that Poisson has positive skewness (long tail towards +x) and positive kurtosis.
